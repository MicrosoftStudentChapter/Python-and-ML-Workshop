```python
plt.style.available
```

```python
plt.plot(x,y,label = 'this is original')
#if want to show more thank one plot
plt.plot(x,x**2 + 7,label = "parabola1")
plt.plot(x,x**2 + 8,label = "parabola2")

#IF WANT TO LABEL
# plt.plot(x,y,label = "a parabola")
#ALSO REQUIRES



plt.title('this is a parabola')
plt.xlabel('x axis')
plt.ylabel('y axis')


plt.legend()







plt.show()
#importtant to be in same line these commands
#might not show in some
```

```python
plt.scatter(x,y,label = 'this is original')
#if want to show more thank one plot
plt.scatter(x,x**2 + 7,label = "parabola1")
plt.scatter(x,x**2 + 8,label = "parabola2")


plt.title('this is a parabola')
plt.xlabel('x axis')
plt.ylabel('y axis')


plt.legend()
#removes the matplotlib label
plt.show()
```

```python
import cv2
```

```python
img = cv2.imread('computer.jpeg')
#comes out as 3d array
```

```python
img.shape
```

```python
#to show image
plt.imshow(img)
#not the real image
 
```

```python
rgb_image = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(rgb_image)
#could also make it black and white
```

```python
gray_image = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(gray_image)
plt.axis("off")
```

```python
#CROP IMAGE

plt.imshow(gray_image)
crop_image = rgb_image[50:175,0:190,:]
plt.imshow(crop_image)
plt.axis("off")
```

```python
crop_image.shape
```

CROPING LARGE IMAGE TO SMALL SIZE

```python
total_pixels = crop_image.shape[0]*crop_image.shape[1]
```

```python
total_pixels
```

```python
#each pixel has 3 values ie RGB
#ie 3 integers are required for each 
```

RESIZING AN IMAGE


```python
crop_image = cv2.resize(crop_image,(100,100))
```

```python
#Lower resolution
plt.imshow(crop_image)
plt.axis("off")
crop_image.shape
```
