```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
from sklearn.datasets import make_blobs
```

```python
X,Y = make_blobs(n_samples = 400,n_features = 2,centers=2)
```

```python
plt.style.use("classic")
plt.scatter(X[:,0],X[:,1],c = Y)
test_point = np.array([0,-3.5])
plt.scatter(test_point[0],test_point[1],color = 'green',marker = '*')
plt.show()
#two classes in this graph class 0 and class 1
```

```python
def distance(a1,a2):
    return np.sum((a1-a2)**2)**.5

```

```python
#KNN(X,Y,test_point)
```

```python
def KNN(X,Y,test_point, k = 5):
    m = X.shape[0]
    vals = []
    for i in range(m):
        d = distance(test_point,X[i])
        vals.append((d,Y[i])) #storing the distance and label
        
    vals = sorted(vals)   
    vals = np.array(vals[:k])  #take first five shortest distance and their labels
    print(vals)
   #print(vals[:,1])
    #only labels
#     print(np.unique(vals[:,1],return_counts = True))
    b = np.unique(vals[:,1],return_counts = True)
    idx = np.argmax(b[1])
    pred = b[0][idx]
    
    return int(pred)
```
